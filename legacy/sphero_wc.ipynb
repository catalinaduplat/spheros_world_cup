```python
import cv2
import numpy as np

# Load the image
image = cv2.imread("Boxes.png")

# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Find the contours of the object
contours, hierarchy = cv2.findContours(gray_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Find the largest contour
c = max(contours, key=cv2.contourArea)

# Get the object's position
(x, y), radius = cv2.minEnclosingCircle(c)

# Draw the object's position on the image
cv2.circle(image, (x, y), radius, (0, 255, 0), 2)

# Display the image
cv2.imshow("Image", image)
cv2.waitKey(0)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create a list of lists
points = [[1, 2], [3, 4], [5, 6]]

# Convert the list of lists to a NumPy array
points_array = np.array(points)

# Extract the x and y coordinates
x = points_array[:, 0]
y = points_array[:, 1]

# Plot the points as scatter
plt.scatter(x, y, marker="x")

# Add the coordinates to the plot
for i, (x_i, y_i) in enumerate(points):
    if i < 2:
        plt.annotate(f"({x_i}, {y_i})", (x_i, y_i), ha="center", textcoords="offset points",
                    xytext=(10, 10), color="red")
    else:
        plt.annotate(f"({x_i}, {y_i})", (x_i, y_i), ha="center")

# Set the title and labels
plt.title("Points with Coordinates")
plt.xlabel("X")
plt.ylabel("Y")

# Invert the y-axis
plt.gca().invert_yaxis()

# Show the plot
plt.show()
```
